```python
# Fix all list-wrapped scalar values in the entire DataFrame
import pandas as pd
import numpy as np
import pandas as pd

def unwrap_any_singleton(cell):
    # Handle both Python lists and NumPy arrays
    if isinstance(cell, (list, np.ndarray)) and len(cell) == 1:
        return cell[0]
    return cell


df_metrics_pet = pd.read_csv("kuay_pet.csv")
df_metrics_pre = pd.read_csv("kuay_pre.csv")
df_metrics_spei = pd.read_csv("kuay_spei.csv")


df_metrics_pet = df_metrics_pet.applymap(unwrap_any_singleton)
df_metrics_pre = df_metrics_pre.applymap(unwrap_any_singleton)
df_metrics_spei = df_metrics_spei.applymap(unwrap_any_singleton)
```

```python
df_metrics_pet.drop(columns=['Unnamed: 0'], inplace=True)
df_metrics_pre.drop(columns=['Unnamed: 0'], inplace=True)
df_metrics_spei.drop(columns=['Unnamed: 0'], inplace=True)
```
